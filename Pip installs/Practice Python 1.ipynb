## Regular Expression

```python
import re

NameAge='''
Ashu is 23 and Manan is 26
Raju is 45 and Tom is 101 and Ron is 1024'''
ages=re.findall(r'\d{1,3}',NameAge) #1,3 a=is the range of digits
names=re.findall(r'[A-Z][a-z]*',NameAge) # [A-Z] means words should start with capital letters,[a-z] for small and * to make it continuous
print(ages)
print(names)

ageDict={}
x=0
for eachname in names:
    ageDict[eachname]=ages[x]
    x+=1
print(ageDict)
```

## face detection


```python
import cv2

# Create a CascadeClassifier Object
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml") #file provided by opensv to identify all face features

# Reading the image as it is
img = cv2.imread("photo.jpg")

# Reading the image as gray scale image
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #cvto color to COLOR_BGR2GRAYis to convert colored image to grayscale

# Search the co-ordintes of the image
faces = face_cascade.detectMultiScale(gray_img, scaleFactor = 1.05,
                                      minNeighbors=5)             #scalefactor dec the shape value by 5%(1.05) till face is found
print(type(faces))
print(faces)

for x,y,w,h in faces:
    img = cv2.rectangle(img, (x,y), (x+w,y+h),(0,255,0),3)  # to find face cordinates and (0,255,0) this gives green rectangle 

resized = cv2.resize(img, (int(img.shape[1]/7),int(img.shape[0]/7)))

cv2.imshow("Gray", img)

cv2.waitKey(0)

cv2.destroyAllWindows()
```

## Video Cordinates
