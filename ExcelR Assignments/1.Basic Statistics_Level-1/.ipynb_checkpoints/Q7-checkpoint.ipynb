```python
import pandas as pd
```

```python
df=pd.read_csv("Q7.csv")
df.head()
```

```python
df.describe()
#gives most of the required values
```

```python
round(df.mean(),1)
#avg points are 3.6, avg score is 3.2 and avg weight is 17.8
```

```python
df.mode()
#there is nothing in common to give a mode 
```

```python
round(df.median(),1)
#almost similar values to mean it means that there is no outlier
```

```python
round(df.var(),2)
#we use variance to see how individual numbers relate to each other within a data set
```

```python
round(df.std(),2)
#we use standard variance to see how close is it to the mean, close it is more reliable it is
#std of weight is the farthest that means that weight data has huge gaps between each data value
```

```python
range1=df.Points.max()-df.Points.min()
range2=df.Score.max()-df.Score.min()
range3=df.Weigh.max()-df.Weigh.min()
print(range1,"\t",range2,"\t",range3)
#range isn't that reliable its better to use IQR instead for accurate results
#like here weight is farthest from the real value
```
