# Importing Libraries

```python
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv('movies.csv')
```

```python
movies.head()
```

```python
movies.shape
```

```python
pd.set_option("Max_columns",None)
```

```python
movies.columns
```

```python
features = ['genres','keywords','tagline','cast','director']
```

```python
movies[features] = movies[features].fillna('')
```

```python
movies[features].isnull().sum()
```

```python
concatenated_features = movies['genres']+' '+movies['keywords']+' '+movies['tagline']+' '+movies['cast']+' '+movies['director']
```

```python
concatenated_features
```

```python
vectorizer = TfidfVectorizer()
vector = vectorizer.fit_transform(concatenated_features)
```

```python
print(vector)
```

```python
cosine = cosine_similarity(vector)
```

```python
print(cosine)
```

```python
cosine.shape
```

```python
movie_name = input(' Enter your favourite movie name : ')
```

```python
titles = movies['title']
print(titles)
```

```python
nearest_match = difflib.get_close_matches(movie_name, titles)
print(nearest_match)
```

```python
nearest_match = nearest_match[0]
print(nearest_match)
```

```python
index_of_movie = movies[movies.title == nearest_match]['index'].values[0]
print(index_of_movie)
```

```python
similarity_score = list(enumerate(cosine[index_of_movie]))
print(similarity_score)
```

```python
sorted_similarity_score = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similarity_score)
```

```python
print('Top 10 Movie recommendations: \n')
i=1
for indexx in sorted_similarity_score:
    indexx= indexx[0]
    title_by_index=movies[movies.index==indexx]['title'].values[0]
    if (i<11):
        print(i,'-',title_by_index)
        i+=1
```

## Prediction System

```python
movie_name = str.capitalize(input(' Enter your favourite movie name : '))
print(movie_name)
titles = movies['title']
nearest_match = difflib.get_close_matches(movie_name, titles)
nearest_match = nearest_match[0]
index_of_movie = movies[movies.title == nearest_match]['index'].values[0]
similarity_score = list(enumerate(cosine[index_of_movie]))
sorted_similarity_score = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print('Top 10 Movie recommendations: \n')
i=1
for indexx in sorted_similarity_score:
    indexx= indexx[0]
    title_by_index=movies[movies.index==indexx]['title'].values[0]
    if (i<11):
        print(i,'-',title_by_index)
        i+=1
```
