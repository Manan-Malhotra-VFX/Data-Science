#### Building Association Rules for France

```python
online_retail_data['Country'].unique()
```

```python
online_retail_data['Country'].nunique()
```

```python
pd.set_option('max')
```

```python
basket_France = (online_retail_data[online_retail_data['Country'] =="France"].groupby(['InvoiceNo', 'Description'])['Quantity'] 
                .sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket_France.head(10)

#Pivot table
```

```python
basket_France.shape
```

```python
#Pivot table
france_data = online_retail_data[online_retail_data['Country']=='France']
france_data = pd.pivot_table(values='Quantity',index='InvoiceNo',columns='Description',data=france_data,).fillna(0)
france_data
```

```python
def encode(x):
    if x<=0:
        return 0
    else:
        return 1
```

```python
france_encoded = france_data.applymap(encode)
france_encoded
```

## 5. Model Building

```python
#Purpose of Apriori : to build freq item sets
freq_itemsets = apriori(df = france_encoded,min_support=0.10,use_colnames=True)
freq_itemsets
```

```python
#Purpose of Association Rules - to generate the best associations
rules= association_rules(df =freq_itemsets ,metric='confidence',min_threshold=0.1)
rules.sort_values(by = 'support',axis=0,ascending=False)
```
